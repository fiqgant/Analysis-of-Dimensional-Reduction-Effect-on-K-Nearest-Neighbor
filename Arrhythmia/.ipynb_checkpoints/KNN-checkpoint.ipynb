```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lux
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

# KNNTanpa Reduksi

```python
df = pd.read_csv('dataset/arrhythmia.csv')
```

```python
df.isnull().sum()
print(df.isnull().sum())
```

```python
# iterating the columns 
for col in df.columns: 
    print(col)
```

```python
from sklearn.preprocessing import StandardScaler

X = df.iloc[:, 0:279].values
x = df.iloc[:, 0:279]
y = df.iloc[:, 279].values
y_name = df.iloc[:, 279:280]
print("X shape:",X.shape)
print("y shape:",y_name.shape)
```

```python
# iterating the columns 
print("Attributes:\n")
for col in x.columns: 
    print(col)

print("\n")
print("Class:")
for col in y_name.columns: 
    print(col)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
KNN = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')
KNN.fit(X_train, y_train)
```

```python
y_pred_KNN = KNN.predict(X_test)
```

```python

matrix = plot_confusion_matrix(KNN, X_test, y_test, cmap=plt.cm.Blues)
matrix.ax_.set_title('Confusion Matrix Logistic Reduction Tanpa Reduksi')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.gcf().axes[0].tick_params
plt.gcf().axes[1].tick_params
plt.gcf().set_size_inches(20,12)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef
ac  = accuracy_score(y_test, y_pred_KNN)
pre = precision_score(y_test, y_pred_KNN, average='weighted')
re  = recall_score(y_test, y_pred_KNN, average='weighted')
f1  = f1_score(y_test, y_pred_KNN, average='weighted')
mc  = matthews_corrcoef(y_test, y_pred_KNN)  
```

```python
np.set_printoptions()
print("Akurasi    =", ac)
print("Precison    =",pre)
print("Recall      =", re)
print("f1 score    =", f1)
print("MCC         =", mc)
```

# KNNREDUKSI LDA

```python
LDA_d3 = pd.read_csv('dataset/arrhythmia_after_LDA.csv')
X = LDA_d3.iloc[:, 0:3].values
x = LDA_d3.iloc[:, 0:3]
y = LDA_d3.iloc[:, 3].values
y_name = LDA_d3.iloc[:, 3:4]
print("X shape:",X.shape)
print("y shape:",y_name.shape)

# iterating the columns 
print("Attributes:\n")
for col in x.columns: 
    print(col)

print("\n")
print("Class:")
for col in y_name.columns: 
    print(col)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
LDA_KNN3D = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')
LDA_KNN3D.fit(X_train, y_train)
y_pred_LDA_KNN3D = LDA_KNN3D.predict(X_test)
```

```python
matrix = plot_confusion_matrix(LDA_KNN3D, X_test, y_test, cmap=plt.cm.Blues)
matrix.ax_.set_title('Confusion Matrix 3 Dimensi')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.gcf().axes[0].tick_params
plt.gcf().axes[1].tick_params
plt.gcf().set_size_inches(20,12)
plt.show()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef
ac  = accuracy_score(y_test, y_pred_LDA_KNN3D)
pre = precision_score(y_test, y_pred_LDA_KNN3D, average='weighted')
re  = recall_score(y_test, y_pred_LDA_KNN3D, average='weighted')
f1  = f1_score(y_test, y_pred_LDA_KNN3D, average='weighted')
mc  = matthews_corrcoef(y_test, y_pred_LDA_KNN3D)  
np.set_printoptions()
print("Akurasi    =", ac)
print("Precison    =",pre)
print("Recall      =", re)
print("f1 score    =", f1)
print("MCC         =", mc)
```
